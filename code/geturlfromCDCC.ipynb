```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import numpy as np
from time import sleep
from time import time
from random import randint
import re
import csv
```

```python
##check the web page
#check the response
url="https://awards.data-viz.cn/2021/view/prize.html"
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
soup.find_all('p', class_="prize-title")
```

```python
award_categorys = []
names = []
prizes = []
links = []
```

```python
infos = soup.find_all('ul', attrs={'class':'prize-panel'})
i = 0

for info in infos: 
    type_ = info.get("data-type")
    print(i)
    #print(info.find_all("li",{"class":True}))
    classes = info.find_all('li')
    for cls in classes:
        a = cls.get("class")
        print(a[0])
        #prizes.append(a[0])
        
    items = info.find_all('div', class_='work-name')
    for item in items:
        if item.get_text() != "空缺":
            print(type_)
            #print(item.get("class"))
            print(item.get_text())
            print(item.find('a').attrs['href'])
    i+=1
    if i == 10:
        break
    
    
```

```python
infos = soup.find_all('ul', attrs={'class':'prize-panel'})
i = 0

for info in infos: 
    type_ = info.get("data-type")
    classes = info.find_all('li')
    for cls in classes:
        a = cls.get("class")
        prizes.append(a[0])
        
    items = info.find_all('div', class_='work-name')
    for item in items:
        if item.get_text() != "空缺":
            award_categorys.append(type_)
            names.append(item.get_text())
            links.append(item.find('a').attrs['href'])
    i+=1
    if i == 10:
        break
    
```

```python
china_data_content_competition2021={
    
    'award_category':award_categorys,
    'prize':prizes,
    'title':names, 
    'link':links
}

dataframe = pd.DataFrame.from_dict(china_data_content_competition2021,orient ='index').transpose()
#check the dataframe
print(dataframe.info())
```

```python
# drop the data with na
dataframe = dataframe.dropna()
dataframe
```
